```python
# Filling missing scores with the mean score
merged_data['score'].fillna(merged_data['score'].mean(), inplace=True)
# Filling missing 'imd_band' with the mode
merged_data['imd_band'].fillna(merged_data['imd_band'].mode()[0], inplace=True)
# For 'date_unregistration', we fill it with a placeholder date if missing
merged_data['date_unregistration'].fillna('9999-12-31', inplace=True)
```

```python
# Fill missing values for assessments columns
merged_data['assessment_type'].fillna('Unknown', inplace=True)  # Placeholder value
merged_data['date'].fillna(merged_data['date'].mean(), inplace=True)  # Fill with mean
merged_data['weight'].fillna(merged_data['weight'].mode()[0], inplace=True)  # Fill with mode
```

```python


# Ensure 'date_unregistration' is treated as a string to handle out-of-bounds values
merged_data['date_unregistration'] = merged_data['date_unregistration'].astype(str)

# Identify and replace out-of-bounds dates
out_of_bounds_dates = merged_data['date_unregistration'] == '9999-12-31'
merged_data.loc[out_of_bounds_dates, 'date_unregistration'] = None

# Handle missing values
# Filling missing scores with the mean score
merged_data['score'].fillna(merged_data['score'].mean(), inplace=True)
# Filling missing 'imd_band' with the mode
merged_data['imd_band'].fillna(merged_data['imd_band'].mode()[0], inplace=True)
# Filling missing 'date_unregistration' with the mode (after ensuring no out-of-bounds dates)
merged_data['date_unregistration'].fillna(merged_data['date_unregistration'].mode()[0], inplace=True)

# Fill missing 'date_registration' with the mode of the column
mode_date_registration = merged_data['date_registration'].mode()[0]
merged_data['date_registration'].fillna(mode_date_registration, inplace=True)

# Fill missing values for assessment-related columns
merged_data['id_assessment'].fillna(-1, inplace=True)  # Placeholder value
merged_data['date_submitted'].fillna(merged_data['date_submitted'].mean(), inplace=True)  # Fill with mean
merged_data['is_banked'].fillna(merged_data['is_banked'].mode()[0], inplace=True)  # Fill with mode

# Fill missing values for assessments columns
merged_data['assessment_type'].fillna('Unknown', inplace=True)  # Placeholder value
merged_data['date'].fillna(merged_data['date'].mean(), inplace=True)  # Fill with mean
merged_data['weight'].fillna(merged_data['weight'].mode()[0], inplace=True)  # Fill with mode

# Convert 'date_unregistration' and 'date_registration' to datetime
merged_data['date_unregistration'] = pd.to_datetime(merged_data['date_unregistration'], errors='coerce')
merged_data['date_registration'] = pd.to_datetime(merged_data['date_registration'], errors='coerce')

# Display the first few rows to verify the conversion
print("First few rows after converting 'date_unregistration' and 'date_registration' to datetime:")
print(merged_data.head())
```

```python
# Detect and handle outliers using IQR method
```

```python
# Create the 'average_score' column
merged_data['average_score'] = merged_data.groupby('id_student')['score'].transform('mean')

# Define numerical columns
numerical_columns = ['score', 'date_submitted', 'date', 'weight', 'average_score']

# Detect and handle outliers using IQR method
for column in numerical_columns:
    if column in merged_data.columns:  # Ensure the column exists
        Q1 = merged_data[column].quantile(0.25)
        Q3 = merged_data[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        merged_data = merged_data[(merged_data[column] >= lower_bound) & (merged_data[column] <= upper_bound)]

print("Data shape after removing outliers:")
print(merged_data.shape)

# Plot the pairplot
sns.pairplot(merged_data[numerical_columns + ['final_result']], hue='final_result', diag_kind='kde')
plt.show()
```
